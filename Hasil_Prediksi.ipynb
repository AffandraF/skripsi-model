```python
import os
import shutil
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
source_dir = '/content/drive/MyDrive/_SKRIPSI/dataset_final'
dest_dir = '/content/dataset_final'
model_dir = '/content/drive/MyDrive/_SKRIPSI/model.keras'

if not os.path.exists(dest_dir):
    shutil.copytree(source_dir, dest_dir)

data_dir = dest_dir
test_dir = os.path.join(data_dir, 'test')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python
model = load_model(model_dir)
```

```python
# Prediksi seluruh data
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)

# Ambil label asli
true_classes = test_generator.classes

# Ambil nama file gambar
filenames = test_generator.filenames
```

```python
# Mapping indeks ke nama kelas
index_to_classname = {index: class_name for class_name, index in
                      test_generator.class_indices.items()}

# Ubah label numerik ke nama kelas sebenarnya
actual_labels = [index_to_classname[i] for i in true_classes]
predicted_labels = [index_to_classname[i] for i in predicted_classes]
```

```python
# Buat DataFrame
df = pd.DataFrame({
    'Nama Gambar': filenames,
    'Aktual': actual_labels,
    'Prediksi': predicted_labels
})
```
